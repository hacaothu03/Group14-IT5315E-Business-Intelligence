# Module 4 - Feature Engineering

Uses Module 3 cleaned data from `CleanData/` and calls reusable code from `src/house_price/`.

```python
from pathlib import Path
import sys

CWD = Path.cwd()
if CWD.name == 'notebooks':
    ROOT = CWD.parent
elif (CWD / 'Module4_5_Modeling').exists():
    ROOT = CWD / 'Module4_5_Modeling'
else:
    ROOT = CWD
sys.path.insert(0, str(ROOT / 'src'))

from house_price.data import load_training_data, load_test_data
from house_price.features import get_engineered_feature_frame
from house_price import config

train_df, train_path = load_training_data()
test_df, test_path = load_test_data()
train_path, test_path, train_df.shape, test_df.shape
```

```python
features_full = get_engineered_feature_frame(
    train_df,
    include_derived=True,
    include_context_features=True,
    include_days_on_market=True,
    include_distance_to_center=True,
    include_neighborhood=True,
    include_time_features=True,
    include_mortgage_rate=True,
    feature_set='full',
)
features_reduced = get_engineered_feature_frame(
    train_df,
    include_derived=True,
    include_context_features=True,
    include_days_on_market=True,
    include_distance_to_center=True,
    include_neighborhood=True,
    include_time_features=True,
    include_mortgage_rate=True,
    feature_set='reduced_linear',
)
features_full.shape, features_reduced.shape
```

```python
config.OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
features_full.head(0).T.reset_index().rename(columns={'index': 'feature'}).to_csv(config.OUTPUT_DIR / 'feature_list_full.csv', index=False)
features_reduced.head(0).T.reset_index().rename(columns={'index': 'feature'}).to_csv(config.OUTPUT_DIR / 'feature_list_reduced_linear.csv', index=False)
```
