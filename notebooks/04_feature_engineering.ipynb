# Module 4 - Feature Engineering

This notebook calls the reusable pipeline under `src/house_price/` instead of duplicating feature logic.

```python
from pathlib import Path
import sys

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT / 'src'))

from house_price.data import load_training_data
from house_price.features import get_engineered_feature_frame
from house_price import config

train_df, resolved = load_training_data(allow_raw_fallback=False)
resolved
```

```python
features_full = get_engineered_feature_frame(
    train_df,
    include_derived=True,
    include_context_features=False,
    feature_set='full',
)
features_reduced = get_engineered_feature_frame(
    train_df,
    include_derived=True,
    include_context_features=False,
    feature_set='reduced_linear',
)
features_full.shape, features_reduced.shape
```

```python
config.OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
features_full.head(0).T.reset_index().rename(columns={'index': 'feature'}).to_csv(config.OUTPUT_DIR / 'feature_list_full.csv', index=False)
features_reduced.head(0).T.reset_index().rename(columns={'index': 'feature'}).to_csv(config.OUTPUT_DIR / 'feature_list_reduced_linear.csv', index=False)
```
